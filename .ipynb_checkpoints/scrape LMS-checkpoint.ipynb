```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys

driver = webdriver.Chrome()
driver.get("https://lms.ashoka.edu.in/Contents/Dashboard.aspx")
assert "Python" in driver.title
elem = driver.find_element_by_name("q")
elem.clear()
elem.send_keys("pycon")
elem.send_keys(Keys.RETURN)
assert "No results found." not in driver.page_source
driver.close()
```

```python
dict = {}
list_of_names = []
list_of_images = []
for i in range(0,65):
    for j in range(0, 6):
        #print("i is ", i, "j is ", j, '\n')
        try:
            element = driver.find_element_by_id('ContentPlaceHolder1_lvStudents_ctrl'+str(i)+'_lblName_'+str(6*i+j))
            list_of_names.append(element.text)
            element_image = driver.find_element_by_id('ContentPlaceHolder1_lvStudents_ctrl'+str(i)+'_imgStudent_'+str(6*i+j))
            print('ContentPlaceHolder1_lvStudents_ctrl'+str(i)+'_lblName_'+str(6*i+j))
            print(element.text, element_image.get_attribute('src'))
            
            list_of_images.append(element_image.get_attribute('src'))
            dict[element.text] = element_image.get_attribute('src')
        except:
            continue
```
